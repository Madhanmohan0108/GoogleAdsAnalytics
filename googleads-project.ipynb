```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/kaggle/input/google-ads-sales-dataset/GoogleAds_DataAnalytics_Sales_Uncleaned.csv")
```

```python
col_g=["Clicks", "Impressions", "Leads", "Conversions", "Conversion Rate"]
for col in col_g:
    df[col]=df[col].fillna(df[col].mean())
```

```python
df.isnull().sum()
```

```python
df['Cost'] = df['Cost'].replace('[\$,]', '', regex=True).astype(float)
df['Cost']=df['Cost'].fillna(df['Cost'].mean())
```

```python
df['Sale_Amount'] = df['Sale_Amount'].replace('[\$,]', '', regex=True).astype(float)
df['Sale_Amount']=df['Sale_Amount'].fillna(df['Sale_Amount'].mean())
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df['Ad_Date']=pd.to_datetime(df['Ad_Date'], errors='coerce')
```

```python
df
```

```python
df.isnull().sum()
```

```python
df['Ad_Date']=df['Ad_Date'].fillna(df['Ad_Date'].mode()[0])
```

```python
df.isnull().sum()
```

```python
coll_g=['Campaign_Name', 'Location', 'Device']
for colll in coll_g:
    df[colll]=df[colll].str.strip().str.lower()
```

```python
df['Location'].unique()
```

```python
df['Device'].unique()
```

```python
fixes={'hyderbad':'hyderabad','hydrabad':'hyderabad', 'hydrebad':'hyderabad', 'hyderabad':'hyderabad'}
df['Location']=df['Location'].replace(fixes)
```

```python
fixes1={'dataanalyticscourse': 'data analytics course', 'data anlytics corse': 'data analytics course', 'data analytics corse': 'data analytics course'}
df['Campaign_Name']=df['Campaign_Name'].replace(fixes1)
```

```python
plt.figure(figsize=((8,5)))
sns.histplot(data=df,x='Clicks',hue='Device',kde=True)
plt.title("Clicks in the devices")
plt.show()
```

```python
plt.figure(figsize=((8,5)))
sns.boxplot(data=df, x='Device', y='Cost')
plt.title("Cost distribution of device")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x='Campaign_Name', y='Clicks', data=df)
plt.title("Clicks Distribution by Campaign")
plt.xlabel("Campaign Name")
plt.ylabel("Clicks")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x='Location', y='Sale_Amount', data=df)
plt.title("Sales amount Distribution by Location")
plt.xlabel("Location")
plt.ylabel("Sale Amount")
plt.xticks(rotation=45)
plt.show()
```

```python
outlier_value = df['Sale_Amount'].quantile(0.99)
outliers = df[df['Sale_Amount'] > outlier_value]
print(outliers)
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x='Ad_Date', y='Clicks', data=df)
```

```python
df['weekday'] = df['Ad_Date'].dt.day_name()
```
